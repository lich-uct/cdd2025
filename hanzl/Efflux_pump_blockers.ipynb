# Efflux pump blockers

Explore drug-like small molecules possibly blocking efflux pumps in human cells. Such molecules have potential to support efficacy of other anti-cancer drugs used in parallel. The cancer cells tend to develop resistance to the primary anti-cancer drug via adaptive expression of efflux pumps which get the drug out of the cell. Blocking these pumps by a second drug used in parallel can thereby improve efficacy of the primary drug.

There are several publickly known binders usable to block the efflux pumps to a certain degree. In addition, we may explore additional new binders based on wet lab proposals. New molecules can be proposed in-silico.
